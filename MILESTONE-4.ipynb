## Question1: What is the relation between Mother’s occupation and Final grade?

```python
q1 = df[["Mother's education", 
         "Father's education", 
         'Expected Cumulative grade point average in the graduation (/4.00)']]
```

```python
# Count occurrences grouped by both columns
grouped_counts = q1.groupby(["Mother's education", 'Expected Cumulative grade point average in the graduation (/4.00)']).size().unstack(fill_value=0)

# Define colors for categories
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728']  # Adjust colors as needed

# Plot stacked bar chart
ax = grouped_counts.plot(kind='bar', stacked=True, figsize=(15, 10), color=colors)

# Add value labels to the bars
for container in ax.containers:
    ax.bar_label(container, label_type='edge')

plt.title("Stacked Bar Graph of Mother's Education Counts, Divided by Final Grade")
plt.xlabel("Mother's education")
plt.ylabel("Counts")
plt.legend(title='Expected Cumulative grade point average in the graduation (/4.00)')
plt.xticks(rotation=0)
plt.show()
```

### From above graph we can see that a lot of Mothers who have only completed primary school have more number of children scoring between 3.00-3.49.
### While Mothers with phd have its children under GPA <2.00

```python
# Count occurrences grouped by both columns
grouped_counts = q1.groupby(["Father's education", 'Expected Cumulative grade point average in the graduation (/4.00)']).size().unstack(fill_value=0)

# Define colors for categories
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728']  # Adjust colors as needed

# Plot stacked bar chart
ax = grouped_counts.plot(kind='bar', stacked=True, figsize=(15, 10), color=colors)

# Add value labels to the bars
for container in ax.containers:
    ax.bar_label(container, label_type='edge')

plt.title("Stacked Bar Graph of Category1 Counts, Divided by Category2")
plt.xlabel("Father's education")
plt.ylabel("Counts")
plt.legend(title='Expected Cumulative grade point average in the graduation (/4.00)')
plt.xticks(rotation=0)
plt.show()
```

# Question 2: What’s the percentage of students commute by different ways have any impact on their salary?

```python
import matplotlib.pyplot as plt
import numpy as np

# Filter dataset
q2 = df[['Transportation to the university', 'Total salary if available']]

# Get unique categories
categories = q2['Transportation to the university'].unique()[:4]  # Select up to 4 categories for 2x2 layout

# Define colors dynamically
color_map = plt.get_cmap('Set3')
colors = [color_map(i % color_map.N) for i in range(len(q2['Total salary if available'].unique()))]

# Create 2x2 subplot layout
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Flatten axes to easily iterate in a loop
axes = axes.flatten()

# Generate pie charts
for ax, category in zip(axes, categories):
    subset_counts = q2[q2['Transportation to the university'] == category]['Total salary if available'].value_counts()
    
    wedges, texts, autotexts = ax.pie(subset_counts, labels=subset_counts.index, autopct='%1.1f%%',
                                      startangle=0, colors=colors[:len(subset_counts)],
                                      textprops={'fontsize': 10, 'weight': 'bold'})  
    
    ax.set_title(f'Distribution of {category}', fontsize=12, weight='bold')

# Adjust layout spacing
plt.tight_layout()
plt.show()
```
